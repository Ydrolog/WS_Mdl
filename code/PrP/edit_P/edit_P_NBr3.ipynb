```python
date_start = '2018-01-01'
date_end = '2019-12-31'
```

```python
f_mult_summer = 0.9
f_mult_winter = 1.1
l_Mo_winter = [10, 11, 12, 1, 2, 3]
l_Mo_summer = [4, 5, 6 , 7, 8, 9]
```

```python
Pa_Fo_P_Src = r"../../../models/NBr/In/CAP/P/NBr1"
Pa_Fo_P_Dst = r"../../../models/NBr/In/CAP/P/NBr3"
```

## 2 Read and prep DF

```python
DF = pd.read_csv(r'../../../data/Dates.csv')
```

```python
DF['Date'] = pd.to_datetime(DF['Date'])
```

```python
DF = DF.loc[(DF['Date'] >= date_start) & (DF['Date'] <= date_end)].reset_index(drop=True)
DF
```

```python
DF['Year'] = DF['Date'].dt.year
DF['Month'] = DF['Date'].dt.month
DF['Day'] = DF['Date'].dt.day
DF['DayOfYear'] = DF['Date'].dt.dayofyear-1
```

## 3. Edit IDF files

### 3.1. Create DF of IDF files, and their months.

```python
os.getcwd()
```

```python
DF = pd.DataFrame()
```

```python
DF = DF = pd.DataFrame({'Fi': [i for i in LD(Pa_Fo_P_Src) if (i.split('_')[1][:4] == '2018')]}) #or (i.split('_')[1][:4] == '2019')] # Contains files to be edited.
DF['year'] = DF['Fi'].apply(lambda x: x.split('_')[1][:4]).astype('int')
DF['Mo'] = DF['Fi'].apply(lambda x: x.split('_')[1][4:6]).astype('int')
```

```python
DF['Mult'] = DF['Mo'].apply(lambda x: f_mult_summer if x in l_Mo_summer else f_mult_winter if x in l_Mo_winter else 1)
```

```python
DF.loc[DF['Mult']==1.1, 'Mo'].unique()
```

```python
DF.loc[DF['Mult']==0.9, 'Mo'].unique()
```

### 3.1 Iterate

```python
for i, row in DF.iterrows():

    metadata, data = fn.read_asc(PJ(Pa_Fo_P_Src, row['Fi'])) # Load the first .asc file

    result = data * row['Mult']

    fn.save_asc(PJ(Pa_Fo_P_Dst, row['Fi']).replace('NBr1', 'NBr3'), metadata, result) # Save the result to a new .asc file

    print(f"{row['Fi']} was processed and saved as {row['Fi'].replace('NBr1', 'NBr3')} in the corresponding folder.")
```
