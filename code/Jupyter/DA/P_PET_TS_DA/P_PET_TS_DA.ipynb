```python
import numpy as np
import os
from os import listdir as LD, makedirs as MDs
from os.path import join as PJ, basename as PBN, dirname as PDN, exists as PE
import matplotlib.pyplot as plt
from tqdm import tqdm  # For progress visualization (optional)
```

# Define functions

```python
def read_asc(file_path):
    """Reads an .asc file and returns the data as a NumPy array."""
    with open(file_path, 'r') as f:
        lines = f.readlines()
    
    # Skip the first 6 lines (header)
    data = np.genfromtxt(lines[6:])
    return data
```

```python

def combine_asc_files(folder_path, file_names):
    """Reads specified .asc files in a folder and combines them into a 3D NumPy array."""
    if not file_names:
        raise ValueError("No .asc files provided.")
    
    arrays = [read_asc(PJ(folder_path, f)) for f in file_names]
    combined_array = np.stack(arrays, axis=0)  # Stack along a new axis
    
    return combined_array
```
