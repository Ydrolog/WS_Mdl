## 0.1. Options

```python
MdlN = 'NBr52'
```

```python
M = Mdl_N(MdlN)
```

```python
MdlN_B = True
if MdlN_B:
    M_B = Mdl_N(M.B)
```

```python
date_B_YMD = '19951231'
date_B = DT.strptime(date_B_YMD, "%Y%m%d")
date_S_YMD = DT.strftime(date_B + timedelta(days=1), "%Y%m%d")
```

# 1. Read HD, select SP.

```python
HDs = imod.mf6.open_hds(M_B.Pa.HD_Out_Bin, M_B.Pa.DIS_GRB, simulation_start_time=DT.strptime(M_B.INI.sdate, "%Y%m%d"))
```

```python
HDs
```

```python
HDs_ = HDs.sel(time=date_B).compute()
```

```python
HDs_.layer.data
```

```python
Pa_Out = M.Pa.In / f'SHD/{MdlN}'
imod.idf.save(Pa_Out, HDs_, pattern=f'SHD_{date_S_YMD}_L{{layer}}_{MdlN}.idf')
```

# 2. Write SHD block

```python
for i in range(37):
    print(fr" 1,2, {i+1:03},   1.000000    ,   0.000000    ,  -999.9900    , '..\..\In\SHD\{MdlN}\SHD_{date_S_YMD}_L{i+1}_{MdlN}.idf' >>> (shd) starting heads (idf) <<<")
```
